````python
import sys, os
sys.path.insert(0, os.path.abspath('../../..'))
from ValaQuenta.modules.sonification import maths as sn
import math
print('sonification module loaded')
print(f'Sample rate: {sn.SAMPLE_RATE} Hz')
print(f'Concert A:   {sn.CONCERT_A} Hz')
````

## Particle Frequency Table

````python
r = sn.particle_frequency_table()
print(f'{'Particle':12s}  {'Freq (Hz)':10s}  {'ω (rad/s)':12s}  Note')
print('-'*55)
for name, data in r['particles'].items():
    freq = float(data['freq_hz'])
    omega = 2 * math.pi * freq
    print(f'{name:12s}  {freq:10.2f}  {omega:12.4f}  {data["note"]}')
````

## Quasi-Particle Rests — Exact Integer Samples

````python
print('Quasi-particle rests (exact sample counts, no floating point):')
for name, samples in [
    ('Phonon',    sn.PHONON_SAMPLES),
    ('Exciton',   sn.EXCITON_SAMPLES),
    ('Magnon',    sn.MAGNON_SAMPLES),
    ('Roton',     sn.ROTON_SAMPLES),
    ('Plasmon',   sn.PLASMON_SAMPLES),
    ('Gravinon',  sn.GRAVINON_SAMPLES),
]:
    duration_ms = 1000 * samples / sn.SAMPLE_RATE
    print(f'  {name:10s}: {samples:6d} samples  ({duration_ms:.2f} ms)')

print(f'\nGravinon: {sn.FIB_NUM}/{sn.FIB_DEN} = φ-convergent Fibonacci ratio')
print(f'φ ≈ {(1+math.sqrt(5))/2:.6f},  144/89 = {144/89:.6f}')
````

## 16 Sedenion Channels = 16 MIDI Channels = 16 Fingers

````python
r = sn.sedenion_channel_map()
print('Sedenion component → MIDI channel → Hand/Side')
print('-'*55)
for ch in r['channels']:
    side = 'LEFT  (octonion base)' if ch['component'] < 8 else 'RIGHT (upper sedenion)'
    print(f"e{ch['component']:2d}  → CH{ch['midi_channel']:2d}  → {side}")
print(f"\nLeft hand  (e₀–e₇):  J_neg / Blue / constraint")
print(f"Right hand (e₈–e₁₅): J_pos / Red  / expansion / life")
print(f"Zero-divisor chords: pairs where eᵢ·eⱼ = 0  — the jazz chord")
````

## Waveform Preview — Higgs Ground State

````python
r = sn.waveform_preview('Higgs', n_samples=256)
print(f"Particle: {r['particle']}")
print(f"Freq:     {r['freq_hz']} Hz")
print(f"ω:        {r['omega_rad_s']:.4f} rad/s")
print(f"Waveform (first 8 samples): {[f'{v:.3f}' for v in r['waveform'][:8]]}")

# ASCII plot
w = r['waveform']
max_w = max(abs(v) for v in w)
print('\nWaveform:')
for v in w[::16]:  # every 16th sample
    bar = int(20 * v / max_w)
    mark = '█' * abs(bar)
    print(f'  {" "*20 if bar < 0 else ""}{mark}')
````
