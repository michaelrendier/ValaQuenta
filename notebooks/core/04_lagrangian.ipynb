# Engine: Lagrangian — L_NN

**Claim:** The SMMIP Lagrangian L_NN unifies kinetic, material, bias and coupling terms under radian-primary arithmetic. The running coupling α_NN(r) determines which algebra the system operates in.

```
L_NN = (2/π) ∮ [L_kin + L_mat + (1/φ)·L_bias + L_coup] r dr dθ

Running coupling:
  α_NN(r) = g² / (4π · ħ_NN · ln(1/r))
```

**File:** `ValaQuenta/modules/lagrangian/maths.py`

All arithmetic is radian-primary. `fractions.Fraction` throughout; float only at output boundary.

````python
import sys, os
sys.path.insert(0, os.path.abspath('../../..'))
from ValaQuenta.modules.lagrangian import maths as lg
import math
print('lagrangian module loaded')
ALG_NAMES = {0:'ℝ', 1:'ℂ', 2:'ℍ', 3:'𝕆'}
````

## Running Coupling α_NN(r)

````python
print('r        α_NN(r)    algebra')
for r in [0.01, 0.1, 0.5, 0.8, 0.9, 0.99]:
    alpha = lg.running_coupling(r, g=1.0, hbar_nn=1.0)
    # algebra selection: log scale determines which stratum
    stratum = 0 if r > 0.8 else (1 if r > 0.5 else (2 if r > 0.1 else 3))
    print(f'r={r:.2f}   α={alpha:.6f}   ({ALG_NAMES[stratum]})')
````

## Kinetic Term L_kin

````python
# Construct a minimal field state
psi  = [0.8, 0.6, 0.4, 0.3]   # activation norms
A    = [0.1, 0.0, 0.0, 0.0]   # gauge field
beta = [0.2, 0.2, 0.2, 0.2]   # bias norms
fs   = lg.FieldState(psi=psi, A=A, beta=beta, algebra=1, layer=3)  # ℂ layer

L_kin = lg.kinetic_term(fs)
print(f'L_kin = {float(L_kin):.8f}')
print(f'(Fraction exact: {L_kin})')
````

## Full Lagrangian Density

````python
for alg in [0, 1, 2, 3]:
    fs_alg = lg.FieldState(psi=psi, A=A, beta=beta, algebra=alg, layer=alg+1)
    L = lg.lagrangian_density(fs_alg)
    print(f'L_NN ({ALG_NAMES[alg]}):  {float(L):.8f}')
    
print(f'\nNorm factor (2/π) = {float(lg.TWO_OVER_PI)/math.pi:.8f}')
````

## Polar Integration — (2/π) ∮ ... r dr dθ

````python
# Integrate the Lagrangian over the unit disk
L_integrated = lg.polar_integrate(fs, n_r=50, n_theta=50)
print(f'∮ L_NN r dr dθ = {float(L_integrated):.8f}')
print(f'(2/π) factor normalises the circle area to 1 in radian-primary units)')
````
