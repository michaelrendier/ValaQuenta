## The (I|O) Map

````python
# inversion_transform: (r, theta) -> (1/r, theta + pi/2)
for r in [0.1, 0.5, 1.0, 2.0, 10.0]:
    r_out, th_out = iv.inversion_transform(r, 0.0)
    print(f"r={r:5.1f} → 1/r={r_out:.4f}   (r·(1/r)={r*r_out:.4f})")  
print(f"\nFixed point: r=1 → {iv.inversion_transform(1.0, 0.0)}")
````

## Involution: (I|O)² = Identity

````python
# Two applications return to start
for r0, th0 in [(0.3, 0.1), (2.0, 0.5), (0.7, 1.2)]:
    r1, th1 = iv.inversion_transform(r0, th0)
    r2, th2 = iv.inversion_transform(r1, th1)
    print(f"(r={r0:.2f}, θ={th0:.2f}) → ({r1:.4f},{th1:.4f}) → ({r2:.4f},{th2:.4f})")
    print(f"  Round-trip error: dr={abs(r2-r0):.2e}, dθ={abs(th2-th0-math.pi):.2e}")
````

## J_N Four-Cycle: Z₄ Group

````python
# Four applications = 2π rotation (return to original angle)
r0, th0 = 2.0, 0.0
print(f"Start: r={r0}, θ={th0:.4f}")
r, th = r0, th0
for i in range(4):
    r, th = iv.inversion_transform(r, th)
    print(f"  Step {i+1}: r={r:.6f}, θ/π={th/math.pi:.4f}")
print(f"\nFour-cycle period = 4×(π/2) = 2π = one full rotation")
print(f"J_N generates Z₄. The group order is 4.")
````
