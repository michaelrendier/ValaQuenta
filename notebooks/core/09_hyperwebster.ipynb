## Horner Bijection — Word → Integer Address

````python
words = ['hello', 'world', 'sigma', 'riemann', 'the']
print(f'{'Word':12s}  {'Horner address':20s}  {'log₁₀(address)'}')
for word in words:
    idx = hw.horner_encode(word)
    print(f'{word:12s}  {idx:20d}  {math.log10(idx+1):.4f}')
print(f'\nBase 97^k: {hw.VOCAB_BASE}^4 = {hw.VOCAB_BASE**4:,} (addresses for 4-char words)')
````

## Fano Address — Octonion Path

````python
for word in ['hello', 'sigma', 'zero']:
    fano_idx = hw.fano_encode(word)
    print(f"'{word}' → Fano indices: {fano_idx}")
print()
print('Fano plane lines (octonion multiplication):')
for i, line in enumerate(hw.FANO_LINES):
    print(f'  Line {i}: points {line} — e{line[0]}·e{line[1]} = e{line[2]}')
````

## SemanticWord — Complete Address

````python
for word in ['hello', 'riemann', 'sigma']:
    sw = hw.SemanticWord(word)
    print(f"\nWord: '{word}'")
    print(f"  Horner:  {sw.horner_index}")
    print(f"  Fano:    {sw.fano_indices}")
    print(f"  Hash:    {sw.prime_hash}")
    print(f"  σ-facet: {sw.sigma_facet:.6f}")
````

## Bijection Verification — Round Trip

````python
# Encode then decode — should recover original word
for word in ['hello', 'test', 'ab', 'z']:
    idx = hw.horner_encode(word)
    recovered = hw.horner_decode(idx, len(word))
    ok = '✓' if recovered == word else '✗'
    print(f"{ok} '{word}' → {idx} → '{recovered}'")
print(f'\nThe HyperWebster is a bijection. Every word has exactly one address.')
print(f'Every valid address maps to exactly one word.')
````
