## Shannon Information Content

````python
# Uniform distribution (max entropy) vs peaked distribution
configurations = [
    ('uniform 8',    [1.0]*8),
    ('peaked',       [0.9, 0.1, 0.05, 0.02, 0.01, 0.01, 0.005, 0.005]),
    ('near-delta',   [0.99, 0.005, 0.002, 0.001, 0.001, 0.0005, 0.0005, 0.0]),
    ('word-like',    [0.7, 0.5, 0.4, 0.3, 0.2, 0.15, 0.1, 0.05]),
]

for name, psi in configurations:
    I = ni.shannon_information(psi)
    print(f'{name:15s}: I_info = {I:.4f} bits')
print(f'\nMax entropy for 8 components: log₂(8) = {math.log2(8):.4f} bits')
````

## Phi Flux Through Algebra Boundary

````python
psi = [0.9, 0.7, 0.5, 0.4, 0.3, 0.2, 0.15, 0.1]
ALG_NAMES = {0:'ℝ', 1:'ℂ', 2:'ℍ', 3:'𝕆'}

print('Phi flux at each algebra boundary:')
for alg in [0, 1, 2, 3]:
    flux = ni.phi_flux(psi, alg)
    print(f'  {ALG_NAMES[alg]}: Φ = {flux:.6f}')
````

## Entropic Time — Where Information Peaks

````python
# Simulate information evolution across layers
layers = []
for layer in range(1, 9):
    # Information grows with layer depth until sedenion boundary
    psi_l = [max(0.01, 1.0 - layer*0.08 + 0.3*math.sin(layer*0.7 + i*0.3)) 
             for i in range(8)]
    I = ni.shannon_information(psi_l)
    flux = ni.phi_flux(psi_l, min(layer//2, 3))
    layers.append((layer, I, flux))
    print(f'Layer {layer}: I_info={I:.4f} bits  Φ={flux:.4f}')

t_e_idx = max(range(len(layers)), key=lambda i: layers[i][1])
print(f'\nt_e = layer {layers[t_e_idx][0]}  (entropic time — information peak)')
````

## Entropic Arrow: ∂_l I_info ≥ 0

````python
print('Information must grow (or stay constant) layer by layer until t_e.')
print('Violation = the system is losing structure — algebra boundary crossed.')
print()

I_prev = 0.0
for layer, I, flux in layers:
    delta = I - I_prev
    arrow = '→' if delta >= 0 else '← VIOLATION'
    print(f'Layer {layer}: ΔI = {delta:+.4f}  {arrow}')
    I_prev = I

print(f'\nNoether conservation law: information only flows forward in layer-time.')
````
